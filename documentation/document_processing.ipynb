```python
from bs4 import BeautifulSoup
import pandas as pd
import os
def extract_text_from_html(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')
        return soup.get_text()
 
    
import markdown2

# Function to extract text from Markdown files
def extract_text_from_md(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        html = markdown2.markdown(text)
        return html
    
import xml.etree.ElementTree as ET

# Function to extract text from XML files
def extract_text_from_xml(file_path):
    tree = ET.parse(file_path)
    root = tree.getroot()
    
    # Extract text from XML elements
    text = ' '.join(elem.text for elem in root.iter() if elem.text)
    return text
```

```python
import os
text_list = []
image_extensions = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp', '.svg', '.ico')
for root, dirs, files in os.walk(os.curdir):
        for file in files:
            if file.lower().endswith(image_extensions):
                file_path = os.path.join(root, file)
                os.remove(file_path)
                print(f"Deleted: {file_path}")
            if file.lower().endswith('.html'):
                 file_path = os.path.join(root, file)
                 extracted_text = extract_text_from_html(file_path)
                 text_list.append(extracted_text)
            if file.lower().endswith('.md'):
                file_path = os.path.join(root, file)
                extracted_text = extract_text_from_md(file_path)
                text_list.append(extracted_text)
            if file.lower().endswith('.xml'):
                file_path = os.path.join(root, file)
                extracted_text = extract_text_from_xml(file_path)
                text_list.append(extracted_text)
                
df = pd.DataFrame({'Text': text_list})
df
```
